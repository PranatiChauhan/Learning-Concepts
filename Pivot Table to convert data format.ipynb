```python
import pandas as pd
d = {"Transaction_Number": [1,1,1,2,2,3,3,3,3,4],
     "Items": ['apples', 'eggs', 'bread','bananas','bread','juice', 
               'ice cream', 'apples','milk','toothpaste'],
     "Quantity": [3,12,1,6,2,1,1,6,2,2]}
df = pd.DataFrame(d)
df
```

How would you convert this to the format you saw earlier?

```python
baskets = df.pivot_table(index = 'Transaction_Number', columns = 'Items',
              values = 'Quantity', aggfunc = 'sum', fill_value = 0)
```

```python
baskets.head()
```

```python
#We need to replace quantities > 0 with 1s
array = baskets.values
array[array > 0] = 1
```

```python
array
```

```python
basket_df = pd.DataFrame(array, columns = baskets.columns)
basket_df
```
